### Проверим информацию о типах данных

```python
df.info()
```

Многие типы данных не удовлетворяют информации, данной для задания. Почти все числа, указанные в датафрейме, имеют тип данных Float (тип данных для чисел с плавающей точкой), но не целочисленный тип. Исправим это с помощью метода astype(). Для начала проверим количество значений с типом None. (astype() не работает для столбцов, содержащих тип None).

Посмотрим все пропущенные значения

```python
print(df.isna().sum())
df = df.dropna(subset = ['salary'])
print(df.isna().sum())
```

В моём случае пропущенных значений было немного, их можно удалить без ущерба для обработки.

```python
df['work_year'] = df['work_year'].astype('int64')
df['salary_in_usd'] = df['salary_in_usd'].astype('int64')
df['salary'] = df['salary'].astype('int64')
```

Для проверки исправления выведем первые 5 колонок

```python
df.head(5)
```

Всё исправлено, идём дальше

т.к. датафрейм небольшой, то с помощью цикла ниже и метода unique() проверить уникальные данные в таблице.

```python
for i in df.columns: 
    print(i,sorted(df[i].unique()), sep = '\n')
    print()
```

При проверке значений было выявлено много некорректных данных, приведём их к общему виду, заменим их на правильные значения

```python
df.loc[:, 'job_title'] = df['job_title'].replace('DataScientist', 'Data Scientist').replace('Data SCIENTIST','Data Scientist').replace('Data AnalyticsManager','Data Analytics Manager')
df.loc[:, 'company_size'] = df['company_size'].replace('Large','L')
```

Оценим данные с помощью метода describe()

Метод describe помогает посмотреть подробную информацию о данных
- count: количество ненулевых значений
- mean: среднее значение
- std: стандартное отклонение
- min: минимальное значение
- 25%: первый квартиль
- 50%: медиана
- 75%: третий квартиль
- max: максимальное значение

```python
df.describe().astype(int)
```

При вызове метода describe() и unique() была обнаружена строка, где значение exp (опыт работы на прошлой работе) равен -1, что быть не может. Удаляем это значение  

```python
df = df[df['exp'] != -1] 
```

Также были выявлены странные данные при проверке датафрейма, выведем их

```python
df[df['company_location'] == 'HU']
```

В данной ячейке имеется очень высокий параметр salary при низком параметре salary_in_usd. Было интересно узнать, сколько существует строк со страной "HU", но данная строка является единственной. Склонюсь к тому, что такое и в правду может быть для данной страны, но лучше удалить эту строку, хоть параметр и единсвенный, это может повлиять на дальнейшие результаты.

```python
df = df[df['company_location'] != 'HU']
```

Опираясь на максимальный возраст разработчка, где всего 1% разработчиков старше 60 лет, а многие работают с 20 лет, то были удалены данные, где общий стаж работы >= 40 годам

https://javarush.com/groups/posts/2763-kofe-breyk-27-opros-razrabotchikov-stack-overflow-2020-jazihki-programmirovanija-obrazovanie-te

```python
df[(df['exp_all'] == 60) | (df['exp_all'] == 81)]
```

```python
df[df['exp_all'] >= 40]
```

```python
df = df[df['exp_all'] < 40]
```

Общий стаж работы не может быть меньше стажа работы на предыдущей работе. Удаляем данные значения

```python
df = df[df['exp_all'] >= df['exp']]
```

теперь можно удалить все явные дубликаты 

```python
print(len(df))
df = df.drop_duplicates().reset_index(drop = True) 
print(len(df))
```

### Матрица диаграммы рассеяния

Для удобства выведем название всех колонок

```python
df.columns
```

Нарисуем матрицу диаграммы рассеяния для числовых столбцов

```python
numeric_cols = ["work_year", "salary", "salary_in_usd", "exp_all", "exp"]
sns.pairplot(df[numeric_cols])
```

Из данной матрицы нужными для работы диаграммами окажутся зависимости между salary от exp_all, salary от salary_in_usd и salary_in_usd от exp_all 

Посчитаем корреляцию и ковариацию для числовых столбцов

```python
correlation_matrix = df[numeric_cols].corr()
covariation_matrix = df[numeric_cols].cov()
```

Выведем тепловую карту корреляции

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=1, vmin = -1)
plt.title("Тепловая карта корреляции")
plt.show()
```

Выводы корреляции:

- Зарплата в долларах (salary_in_usd) сильно коррелирует с общим опытом (exp_all), что подтверждает связь между большим опытом и высокими зарплатами на международных должностях.


- Корреляции между опытом в текущей должности (exp) и другими параметрами очень слабые, что может говорить о том, что опыт на текущей должности не играет ключевую роль в определении зарплаты.


- Связь между годом работы и зарплатой в долларах положительная, что может говорить об увеличении зарплат в последние годы.

По - моему мнению, удобным способом представления ковариации является таблица. Выведем её

```python
covariation_matrix.astype('int64')
```

Выводы ковариации:

- Зарплата в долларах (salary_in_usd) и общий опыт (exp_all) имеют наиболее значительную положительную ковариацию, что указывает на тесную связь между этими переменными. Это значит, что с ростом общего опыта возрастает и зарплата в долларах.


- Слабая ковариация между опытом на текущей должности и зарплатой указывает на то, что фактор опыта в данной роли оказывает меньшее влияние на зарплату, чем общий опыт.


- Связь между годом работы и зарплатой в долларах умеренная и положительная, но она не является доминирующим фактором.

### Task1
Использовать seaborn. По группировке - work_year и количество компаний по
различному типу занятости (employment_type) построить диаграмму следующего вида:


Для начала сгруппируем элементы с помощью groupby()

```python
df_agg = df.groupby(['work_year','employment_type']).size().reset_index(name = 'count')
df_agg
```

### Вывод из 1 задания:

### Task2
Использовать pandas и plot. По сводной таблице (pivot_table) - отобразить среднюю
зарплату в usd по локациям (company_location). Сделать линию зелёной, шириной 4 и отобразить

### Task3
Использовать matplotlib. Построить круговую диаграмму, которая отображает процент
количества записей по каждому году.


Сначала сгруппируем данные и применим метод size() для вывода количества значений.

```python
df_group_work_year = df.groupby(['work_year']).size()
df_group_work_year
```
